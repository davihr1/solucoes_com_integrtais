# 6.2 Regra básica de Integração

**Exemplo 1: Integral de uma função polinomial simples**

```python
import sympy as sp

x = sp.symbols('x')

f = x**3

integral = sp.Integral(f, x)
integral

#Integração Indefinida
integral = sp.integrate(f, x)
integral
```

**Exemplo 2: Função trigonométrica**

```python
y = sp.symbols('y')

#Definindo a função trigonométrica
h = sp.sin(y)
```

```python
#calculando a integral indefinida de sin(x)
integral  = sp.integrate(h, y)

print("A integral de sin(x) é: " )
print(integral)
```

# **6.3 Aplicação do metodo de substituisção**

**Exemplo 1:** integral de 1/x com método da substituição

```python
from sympy import symbols, integrate, Rational, pprint

x, u = symbols('x u')

f = 1 / x
f
```

```python
#Aplicando a Substituição
substituicao = u**2
integral_substituicao = integrate(f.subs(x, substituicao) * 2*u, u)
pprint(integral_substituicao, use_unicode=True);
```
