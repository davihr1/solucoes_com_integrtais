```python
from sympy import symbols, limit, sin, cos, exp, sqrt

#definie a variavel simbolica X
x = symbols('x');

#Funções para cálculo de limites
functions = [
    (x**2 - 4) / (x - 2),   #limite1: (x^2 - 4) / (x - 2)
    sin(x) / x,             #limite2: sin(x) / x
    (cos(x) - 1) / x**2,    #limite3: (cos(x)  -1 ) / X^2
    sqrt(1 + x) - sqrt(1 - x)  #limite4: sqrt(1 + x) - sqrt(1 - x)

]

#calcular os limites e imprimir os resultados
for i, f in enumerate(functions, start=1):
  lim = limit(f, x, 0)
  print(f"limite {i}: {f}, quando x se aproxima de 0 é : {lim}\n")

```

Continuidade de Funções  de uma variavel:

```python
from sympy import symbols, limit, sin, cos, exp, sqrt

x = symbols('x');

functions = [
    sin(x),
    cos(x),
    exp(x),
    1 / (x + 1)
];

ponto = 0;

for f in functions:
  lim_esquerda = limit(f, x, ponto, '-').doit()
  lim_direita = limit(f, x, ponto, '+').doit()

  if f.subs(x, ponto) == lim_esquerda == lim_direita:
    print(f'A função {f} é suave em {ponto}')
  else:
    print(f'A função {f} não é suave em {ponto}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, limit, sin, cos, exp, sqrt

x = np.linspace(-10, 10, 1000)

y_sin = np.sin(x)
y_cos = np.cos(x)
y_exp = np.exp(x)
y_inv = 1 / (x + 1)

plt.figure(figsize=(12, 6))

plt.plot(x, y_sin, label='sin(x)', color='blue')
plt.plot(x, y_cos, label='cos(x)', color='red')
plt.plot(x, y_exp, label='exp(x)', color='green')
plt.plot(x, y_inv, label='1 / (x + 1)', color='purple')

plt.title('Gráficos das Funções')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.ylim(-2, 2)
plt.grid(True)
plt.show()
```

Continuidade de Duas Variaveis:

Esse código verifica a continuidade de uma função bidmensional f(x,y) = x**2y por x**2 + y**2 no ponto (0, 0).

ele calcula o limite da função conforme x e y se aproxima de 0 e verifica se esse limite é igual ao valor da função no ponto (0,0). se forem iguais, é continua; caso contrario, não é continua.

```python
from sympy import symbols, limit

x, y = symbols('x y')

f = x**2 * y / (x**2 + y**2)

ponto = (0, 0)


limite = limit(f, (x, ponto[0]), (y, ponto[1])).doit()


if f.subs({x: ponto[0], y: ponto[1]}) == limite:
  print(f'A função {f} é suave em {ponto[0]} e y {ponto[1]}')
else:
  print(f'A função {f} não é suave em {ponto[0]} e y {ponto[1]}')
```
