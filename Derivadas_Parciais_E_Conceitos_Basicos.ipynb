# 5.1 Definição de Derivada Parcial


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x, y = sp.symbols('x y')

f = x**2 + y**2

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

f_num = sp.lambdify((x, y), f, 'numpy')
df_dx_num = sp.lambdify((x, y), df_dx, 'numpy')
df_dy_num = sp.lambdify((x, y), df_dy, 'numpy')

x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
x_vals, y_vals = np.meshgrid(x_vals, y_vals)

f_vals = f_num(x_vals, y_vals)
df_dx_vals = df_dx_num(x_vals, y_vals)
df_dy_vals = df_dy_num(x_vals, y_vals)

plt.figure(figsize=(18, 5))

# Primeiro gráfico: Função original
plt.subplot(1, 3, 1)
plt.contourf(x_vals, y_vals, f_vals, cmap='viridis')
plt.colorbar(label='f(x, y)')
plt.title('$f(x, y) = x^2 + y^2$')
plt.xlabel('X')
plt.ylabel('Y')

# Segundo gráfico: Derivada parcial em relação a x
plt.subplot(1, 3, 2)
plt.contourf(x_vals, y_vals, df_dx_vals, cmap='viridis')
plt.colorbar(label='$\\frac{\\partial f}{\\partial x} = 2x$')
plt.title('$\\frac{\\partial f}{\\partial x} = 2x$')
plt.xlabel('X')
plt.ylabel('Y')

# Terceiro gráfico: Derivada parcial em relação a y
plt.subplot(1, 3, 3)
plt.contourf(x_vals, y_vals, df_dy_vals, cmap='viridis')
plt.colorbar(label='$\\frac{\\partial f}{\\partial y} = 2y$')
plt.title('$\\frac{\\partial f}{\\partial y} = 2y$')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```

# 5.2 Calculos das Derivadas parciais em Funções Simples

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x, y = sp.symbols('x y')

# f(x,y) = sin(x) * cos(y)
f = sp.sin(x) * sp.cos(y)

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

f_num = sp.lambdify((x, y), f, 'numpy')
df_dx_num = sp.lambdify((x, y), df_dx, 'numpy')
df_dy_num = sp.lambdify((x, y), df_dy, 'numpy')

# Ajuste para usar numpy (np) corretamente
x_vals = np.linspace(-2*np.pi, 2*np.pi, 100)
y_vals = np.linspace(-2*np.pi, 2*np.pi, 100)
x_vals, y_vals = np.meshgrid(x_vals, y_vals)

f_vals = f_num(x_vals, y_vals)
df_dx_vals = df_dx_num(x_vals, y_vals)
df_dy_vals = df_dy_num(x_vals, y_vals)

plt.figure(figsize=(18, 5))

# Gráfico 1: Função original
plt.subplot(1, 3, 1)
plt.contourf(x_vals, y_vals, f_vals, cmap='viridis')
plt.colorbar(label='f(x, y)')
plt.title(r'$f(x, y) = \text{sin}(x) \cdot \text{cos}(y)$')
plt.xlabel('X')
plt.ylabel('Y')

# Gráfico 2: Derivada em relação a X
plt.subplot(1, 3, 2)
plt.contourf(x_vals, y_vals, df_dx_vals, cmap='viridis')
plt.colorbar(label=r'$\frac{\partial f}{\partial x}$')
plt.title(r'$\frac{\partial f}{\partial x} = \text{cos}(x) \cdot \text{cos}(y)$')
plt.xlabel('X')
plt.ylabel('Y')

# Gráfico 3: Derivada em relação a Y
plt.subplot(1, 3, 3)
plt.contourf(x_vals, y_vals, df_dy_vals, cmap='viridis')
plt.colorbar(label=r'$\frac{\partial f}{\partial y}$')
plt.title(r'$\frac{\partial f}{\partial y} = -\text{sin}(x) \cdot \text{sin}(y)$')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```

# 5.3 Interpretação de Derivadas Parciais Graficamente

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Adicionando o prefixo sp. para as funções do SymPy
x, y = sp.symbols('x y')

# Função: Paraboloide Hiperbólico (Sela)
f = x**2 - y**2

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

f_num = sp.lambdify((x, y), f, 'numpy')
df_dx_num = sp.lambdify((x, y), df_dx, 'numpy')
df_dy_num = sp.lambdify((x, y), df_dy, 'numpy')

x_vals = np.linspace(-2, 2, 100)
y_vals = np.linspace(-2, 2, 100)
x_vals, y_vals = np.meshgrid(x_vals, y_vals)

f_vals = f_num(x_vals, y_vals)
df_dx_vals = df_dx_num(x_vals, y_vals)
df_dy_vals = df_dy_num(x_vals, y_vals)

plt.figure(figsize=(18, 5))

# Subplot 1: Função original
plt.subplot(1, 3, 1)
plt.contourf(x_vals, y_vals, f_vals, cmap='viridis')
plt.colorbar(label='f(x, y)')
plt.title(r'$f(x, y) = x^2 - y^2$')
plt.xlabel('X')
plt.ylabel('Y')

# Subplot 2: Derivada em relação a X
plt.subplot(1, 3, 2)
plt.contourf(x_vals, y_vals, df_dx_vals, cmap='viridis')
plt.colorbar(label='df/dx')
plt.title(r'$\frac{\partial f}{\partial x} = 2x$')
plt.xlabel('X')
plt.ylabel('Y')

# Subplot 3: Derivada em relação a Y
plt.subplot(1, 3, 3)
plt.contourf(x_vals, y_vals, df_dy_vals, cmap='viridis')
plt.colorbar(label='df/dy')
plt.title(r'$\frac{\partial f}{\partial y} = -2y$')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```

# 5.4 Derivadas Parciais de Funções Compostas

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x, y = sp.symbols('x y')

f = sp.sin(x**2 + y)

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

f_num = sp.lambdify((x, y), f, 'numpy')
df_dx_num = sp.lambdify((x, y), df_dx, 'numpy')
df_dy_num = sp.lambdify((x, y), df_dy, 'numpy')

# Ajuste de intervalo
x_vals = np.linspace(-3, 3, 100)
y_vals = np.linspace(-3, 3, 100)
x_vals, y_vals = np.meshgrid(x_vals, y_vals)

f_vals  = f_num(x_vals, y_vals)
df_dx_vals = df_dx_num(x_vals, y_vals)
df_dy_vals = df_dy_num(x_vals, y_vals)

plt.figure(figsize=(18, 5))

# Gráfico 1: Função original
plt.subplot(1, 3, 1)
plt.contourf(x_vals, y_vals, f_vals, cmap='viridis')
plt.colorbar()
plt.title(r'$f(x, y) = \text{sin}(x^2 + y)$')
plt.xlabel('X')
plt.ylabel('Y')

# Gráfico 2: Derivada em X
plt.subplot(1, 3, 2)
plt.contourf(x_vals, y_vals, df_dx_vals, cmap='viridis')
plt.colorbar()
plt.title(r'$\frac{\partial f}{\partial x} = 2x \cdot \text{cos}(x^2 + y)$')
# Note o uso de \\ antes de comandos sensíveis como frac e text
plt.xlabel('X')
plt.ylabel('Y')

# Gráfico 3: Derivada em Y
plt.subplot(1, 3, 3)
plt.contourf(x_vals, y_vals, df_dy_vals, cmap='viridis')
plt.colorbar()
plt.title(r'$\frac{\partial f}{\partial y} = \text{cos}(x^2 + y)$')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```
