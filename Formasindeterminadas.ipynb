```python
from sympy import symbols, limit, sin, cos, exp, sqrt

#definie a variavel simbolica X
x = symbols('x');

#Funções para cálculo de limites
functions = [
    (x**2 - 4) / (x - 2),   #limite1: (x^2 - 4) / (x - 2)
    sin(x) / x,             #limite2: sin(x) / x
    (cos(x) - 1) / x**2,    #limite3: (cos(x)  -1 ) / X^2
    sqrt(1 + x) - sqrt(1 - x)  #limite4: sqrt(1 + x) - sqrt(1 - x)

]

#calcular os limites e imprimir os resultados
for i, f in enumerate(functions, start=1):
  lim = limit(f, x, 0)
  print(f"limite {i}: {f}, quando x se aproxima de 0 é : {lim}\n")

```
